```python
import requests
longtitutes = df3['lon'].tolist()
latitutes = df3['lat'].tolist()
url = 'https://geo.fcc.gov/api/census/block/find?'
fips3= []
for lat,long in zip(latitutes,longtitutes):
    print(lat,long)
    parameters = dict(latitude = lat, longitude = long,format = 'xml')
    response = requests.get(url,params=parameters)
    xmlTree = response.content
    fips3.append(xmlTree)


from xml.etree import cElementTree as ET
hope3 =[]
for thing in fips3:
    root = ET.fromstring(thing)
    a = list(root)
    b = a[1].attrib
    hope3.append(b['FIPS'])
```

```python
import ast
westskills = []
westsalary = []
eastskills = []
eastsalary = []
for i in range(len(df)):
    x = ast.literal_eval(df['Skills'][i])
    skill = x[0].split(" ")
    y = ast.literal_eval(df['Job Salary'][i])
    if -125<df['lon'][i]<-110:
        westskills += skill
        westsalary += y
        i +=1
    elif -85<df['lon'][i]<-65:
        eastskills += skill
        eastsalary += y
        i += 1
    else:
        i+=1
    
```
